```python
import pandas as pd
```

```python
# llm pruner style, 0.2 ratio
trainer_state = pd.read_pickle("./tmp/trainer_state_llm_pruner_style.pkl")
```

```python
df = pd.DataFrame(trainer_state.log_history)
```

```python
df = df.set_index("step")[["loss", "eval_loss"]]
```

```python
df["rolling_loss"] = df["loss"].ffill().rolling(10).mean()
```

```python
df["ffill_eval_loss"] = df["eval_loss"].ffill()
```

```python
df[["rolling_loss", "ffill_eval_loss"]].plot()
```
